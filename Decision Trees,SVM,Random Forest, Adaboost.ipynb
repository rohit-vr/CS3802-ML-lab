```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors = 'coerce') 
df['TotalCharges'].dtype
```

For the given data checked for Null values and converted total charges column from object to float


# Data cleaning and processing


```python
df.drop(['customerID'],axis = 1,inplace = True)
```

```python
df
```

ONE HOT ENCODING IS DONE TO CONVERT OBJECT VALUES WITH RESPECTIVE TO TARGET COLUMN

```python
df1=pd.get_dummies(data=df,columns=['gender', 'Partner', 'Dependents', 
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn'], drop_first=True)
```

```python
df1
```

```python
df1.shape
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")

df1.TotalCharges = imputer.fit_transform(df1["TotalCharges"].values.reshape(-1, 1))
```

```python
df1
```

```python
from sklearn.preprocessing import StandardScaler
df2 = StandardScaler()
```

```python
df2.fit(df1.drop(['Churn_Yes'],axis = 1))
features = df2.transform(df1.drop('Churn_Yes',axis = 1))
```

```python
df2
```

Train and test split after feature scaling


```python
from sklearn.model_selection import train_test_split
X =features
Y = df1['Churn_Yes']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state=27)
```

# Predicting using decision tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score ,confusion_matrix

dt = DecisionTreeClassifier()

dt.fit(X_train, Y_train)
y2_predict = dt.predict(X_test)
y2_predict
```

```python
print(classification_report(Y_test,y2_predict))
```

```python
accuracy_score(Y_test,y2_predict)
```

# Predicting using svm 

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, Y_train)
y3_predict = svc.predict(X_test)
```

```python
y3_predict
```

```python
print(classification_report(Y_test,y3_predict))
```

```python
accuracy_score(Y_test,y3_predict)
```

RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=200, random_state=20)

rf_classifier.fit(X_train, Y_train)


y4_pred = rf_classifier.predict(X_test)
y4_pred
```

```python
print(classification_report(Y_test,y4_pred))
```

```python
accuracy_score(Y_test,y4_pred)
```

```python
from sklearn.model_selection import GridSearchCV
param = {'n_estimators':[50,75,100,125,200],
 'criterion':['gini','entropy'],
 'bootstrap':[True, False]}
```

```python
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)


grid_search.fit(X, Y)


print("Best Parameters using Grid Search:", grid_search.best_params_)
print("Best Accuracy Score using Grid Search:", grid_search.best_score_)


```

```python
from sklearn.model_selection import RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=rf_classifier, param_distributions=param, n_iter=10, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)


random_search.fit(X, Y)


print("\nBest Parameters using Randomized Search:", random_search.best_params_)
print("Best Accuracy Score using Randomized Search:", random_search.best_score_)
```

```python
from sklearn.ensemble import AdaBoostClassifier
base_classifier = RandomForestClassifier(n_estimators=200, random_state=20)


adaboost_classifier = AdaBoostClassifier(base_estimator=base_classifier, n_estimators=50, random_state=42)


adaboost_classifier.fit(X_train, Y_train)


y5_pred = adaboost_classifier.predict(X_test)


print(y5_pred)
```
